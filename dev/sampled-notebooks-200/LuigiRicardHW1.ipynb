```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes=True)
```

```python
names = ['L-CORE','L-SURF','L-O2','L-BP','SURF-STBL','CORE-STBL','BP-STBL','COMFORT','ADM-DECS']
post_operative = pd.read_csv('post-operative.csv', names = names)
```

```python
post_operative
```

```python
g=sns.PairGrid(x_vars=["L-CORE","L-SURF","L-O2","L-BP","SURF-STBL","CORE-STBL","BP-STBL","ADM-DECS"], y_vars=["COMFORT"], data =post_operative )
g.map(sns.violinplot, palette="pastel");
```

```python
sns.countplot(x="L-CORE", data = post_operative, order = ['low','mid','high'])
```

```python
sns.countplot(x="L-SURF", data = post_operative, order = ['low','mid','high'])
```

```python
sns.countplot(x="L-BP", data = post_operative, order = ['low','mid','high'])
```

```python
L1_mapping = {'high':3,'mid':2,'low':1}
post_operative['L-CORE'] = post_operative['L-CORE'].map(L1_mapping)
post_operative['L-SURF'] = post_operative['L-SURF'].map(L1_mapping)
post_operative['L-BP'] = post_operative['L-BP'].map(L1_mapping)
LO2_mapping = {'excellent':2,'good':1}
post_operative['L-O2'] = post_operative['L-O2'].map(LO2_mapping)
```

The above code changes the ordinal values of the data into numerical ones.

```python
pd.get_dummies(post_operative)
```
