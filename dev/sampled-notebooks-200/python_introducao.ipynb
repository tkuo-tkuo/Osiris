## Classes

```python
class Retangulo:
    lado_a=None
    lado_b=None
    
    def __init__(self, lado_a, lado_b):
        self.lado_a = lado_a
        self.lado_b = lado_b
        print ('Criando nova instância Retangulo')
    
    def calcula_area(self):
        return self.lado_a * self.lado_b
    
    def calcula_perimetro(self):
        return 2 * self.lado_a + 2*self.lado_b
    
    
```

```python
a=Retangulo(5,8)
```

```python
a
```

```python
a.calcula_area()
```

```python
a.calcula_perimetro()
```

*Heranças*

```python
class Foo:
    a = 1
    def chesse (self):
        print ('queijo')
    
    def foo(self):
        print('comida')
        
class Bar(Foo):
    def bar(self):
        print ('bar')
        
    def foo(self):
        print ('foo de bar!')
```

```python
her=Foo()
```

```python
her.foo()
```

```python
her1=Bar()
```

```python
her1.bar()
```

```python
her1.foo()
```

```python
her1.chesse()
```

*Para acessar os métodos de uma classe usa-se:*

```python
class Bar(Foo):
    def bar(self):
        print ('bar')
        
    def foo(self):
        Foo.foo(self)
        print ('comida de bar!')
```

```python
her1.foo()
```

```python
her1=Bar()
her1.foo()
```
